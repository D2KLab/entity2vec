```python
%pylab inline
from sklearn.decomposition import PCA
import pandas as pd 
import numpy as np
```

```python
#define the dataset and the variables

dataset = 'bgslith'

if dataset == 'mutag':
    y_label = 'label_mutagenic'
elif dataset == 'bgslith':
    y_label = 'label_lithogenesis'
elif dataset == 'bgstheme':
    y_label = 'label_theme'
elif dataset == 'aifb':
    y_label = 'label_affiliation'
elif dataset == 'am':
    y_label = 'label_category'
    
p = 1
q = 4
dimension = 500
model = 'skip'
```

```python
path = '../emb/%s/%d_%s/%s_p%d_q%d_%d_%s.emd' %(dataset,dimension,model,dataset,p,q,dimension,model)

embeddings = pd.read_table(path, skiprows = 1, header = None, sep = ' ')
embeddings.head
```

```python
complete_set = pd.read_table('../datasets/%s/completeDataset.tsv' %dataset)
complete_set.head
```

```python
def convert_labels_to_int(label,dataset):

    if dataset == 'aifb':

        if label == 'http://www.aifb.uni-karlsruhe.de/Forschungsgruppen/viewForschungsgruppeOWL/id1instance':
            return 0

        elif label == 'http://www.aifb.uni-karlsruhe.de/Forschungsgruppen/viewForschungsgruppeOWL/id2instance':
            return 1

        elif label == 'http://www.aifb.uni-karlsruhe.de/Forschungsgruppen/viewForschungsgruppeOWL/id3instance':
            return 2

        else:
            return 3

    elif dataset == 'mutag':
        #already binary values
        return int(label)
    
    elif dataset == 'bgslith':
        if label == 'http://data.bgs.ac.uk/id/Lexicon/LithogeneticType/FLUV'
            return 0
        else:
            return 1
    
    else: #bgslith
        possible_labels = []

        for i in set(complete_set[y_label]):
            possible_labels.append(i)

        labels_to_int = {}

        for i in range(len(possible_labels)):

            labels_to_int[possible_labels[i]] = i

        return labels_to_int[label]
```

```python
def get_embeddings(ID, embeddings):



    embds = embeddings[embeddings[0] == ID].values[0]

    return embds[1:] #the first is the id
```

```python
complete_ids = np.array([int(i) for i in complete_set['id']])

y = np.array([convert_labels_to_int(i, dataset) for i in complete_set[y_label]])

N = len(complete_ids)

X = np.zeros((N,dimension))

for i in range(N):

    emb = get_embeddings(complete_ids[i], embeddings)

    emb = emb.reshape((1,dimension))

    X[i] =  emb
```

```python
target_names = ['no_%s'%(y_label.replace('label_','')),'%s'%(y_label.replace('label_',''))]
#target_names = ['group1','group2','group3','group4']
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
plt.figure(figsize = (20,20))
colors = ['navy', 'red']
lw = 2
for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name, s = 100)
plt.legend(loc='best', shadow=False, scatterpoints=1,fontsize = 40)
plt.title('%s dataset' %dataset, fontdict = {'fontsize' : 50})
plt.savefig('%s_p%d_q%d.png'%(dataset,p,q))
```
